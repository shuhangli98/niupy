```python
import numpy as np
import matplotlib.pyplot as plt
import wicked as w

w.reset_space()
# alpha
w.add_space('c', 'fermion', 'occupied', list('klmn'))
w.add_space('v', 'fermion', 'unoccupied', list('efgh'))
w.add_space('a', 'fermion', 'general', list('oabrstuvwxyz'))
# Beta
w.add_space('C', 'fermion', 'occupied', list('KLMN'))
w.add_space('V', 'fermion', 'unoccupied', list('EFGH'))
w.add_space('A', 'fermion', 'general', list('OABRSTUVWXYZ'))
wt = w.WickTheorem()
```

```python
with open('ee_contractions_analysis.py', 'r') as f:
    rawcode = f.read()

code = rawcode.split('def')[1:]

codeblock = {}
for block in code:
    key = block.split("sigma['")[1].split("']")[0]
    codeblock[key] = block.split('\n')[4:-3]
```

```python
for key in codeblock.keys():
    print(key)
    max_a = 0
    max_a_loc = 0
    max_cv = 0
    max_cv_loc = 0
    for idx, i in enumerate(codeblock[key]):
        scaling = w.analyze_einsum(i, root_index='p')
        scaling_combined = {'a': scaling['a']+scaling['A'], 'c': scaling['c']+scaling['C'], 'v': scaling['v']+scaling['V']}
        if scaling_combined['a'] > max_a:
            max_a = scaling_combined['a']
            max_a_loc = idx
        if scaling_combined['c'] + scaling_combined['v'] > max_cv:
            max_cv = scaling_combined['c'] + scaling_combined['v']
            max_cv_loc = idx
    
    if (max_a > 0): print('Max a:', max_a, codeblock[key][max_a_loc])
    if (max_cv > 0): print('Max cv:', max_cv, codeblock[key][max_cv_loc])
```
